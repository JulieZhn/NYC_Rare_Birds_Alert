```python
import requests
from bs4 import BeautifulSoup
my_url = "https://ebird.org/alert/summary?sid=SN35466"
raw_html = requests.get(my_url).content
```

```python
soup_doc = BeautifulSoup(raw_html, "html.parser")
```

```python
birds = soup_doc.select('div.Observation')
for bird in birds:
    print('----')
    print(bird.text)
```

```python
birds = soup_doc.select('div.Observation')
for bird in birds:
    print('----')
    #print(bird.text)

    name = bird.select_one('.Heading-main').text
    print(name)

    location = bird.select('a')[2].text
    print(location)
    
    link_url = bird.select_one('a[href]')['href']
    print(link_url)

    
```

```python
all_birds = []

birds = soup_doc.select('div.Observation')
for bird in birds:
    print('----')
    #print(bird.text)

    row = {}
    row['name'] = bird.select_one('.Heading-main').text
    #names = bird.select_one('h3.Heading-main').text
    #but if I add .text, it gives me an error message? 
    #It seems work well in the last cell where I also 
    #print(names)

    row['location'] = bird.select('a')[2].text
    
    #link_url = bird.select_one('a[href]')['href']
    row['link_url'] = bird.select_one('a[href]')['href']
    #print(link_url)

    
    print(row)
    
    all_birds.append(row)
    
```

```python
import pandas as pd
df = pd.DataFrame(all_birds)
df.head(20)
```
